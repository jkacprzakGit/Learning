```python
# Reading from file
#open the file and setup the loop
#get number of Cards in file

def Get_number_of_cards():
    file = open("/adshome/jkacprz/TestDay4.txt", "r")
    #file = open("/adshome/jkacprz/AdventOfCodeDay4_input.txt", "r")
    Counter = 0
    Ctr=0

    Content = file.read()
    CoList = Content.split("\n")


    for i in CoList:
        if i:
            Counter += 1
    return(Counter)
```

```python
def Get_num_matches(CardList):
    count=0
    
    for i in CardList: 
        count = count+1
      
    return count
```

```python
def Create_card_list(C): #c is the original card number to genereate the matches and find the next list

    fstr=Get_card_data(C)
    #remove the card information
    s = fstr.find(":")
    p=fstr[0:s+1]
    
    #strip out the card number
    CardNumber=C+1 #index starts at 0

    NewStr=fstr.replace(p, "")

    #remove leading spaces
    NewStr=NewStr.strip() 


    # split into two strings then replace spaces with commas and insert into a list


    Bar = NewStr.find("|")


    #Clean up spaces
    NewStr=NewStr.replace('  ',' ')


    #Find the separator

    Bar = NewStr.find("|")


    #Split into two strings
    PartA = NewStr[0:Bar]

    PartB=NewStr[Bar+1:len(NewStr)]


    #Strip out any leading or trailing spaces
    PartA=PartA.strip() 
    PartB=PartB.strip() 


    #create the lists
    ListA = list(PartA.split(" "))
    ListB = list(PartB.split(" "))


    CardList = []
    count=0
    Pts=0
    CardPts=0
    PrevPts = 0



    for element in ListA:
        if element in ListB:
            if element != '':
                CardList.append(element)
                
    return(CardList)
```
