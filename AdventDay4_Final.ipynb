```python
#open the file and setup the loop
file = open("/adshome/jkacprz/AdventOfCodeDay4_input.txt", "r")
Counter = 0
 
# Reading from file
Content = file.read()
CoList = Content.split("\n")
TotalPts = 0

for i in CoList:
    if i:
        Counter += 1
        

        #read in the next line
        fstr=i
      

        #remove the card information
        s = fstr.find(":")
      
        p=fstr[0:s+1]
        print(p)
      

        NewStr=fstr.replace(p, "")
      
        #remove leading spaces
        NewStr=NewStr.strip() 
      

        # split into two strings then replace spaces with commas and insert into a list

        
        Bar = NewStr.find("|")
      

        #Clean up spaces
        NewStr=NewStr.replace('  ',' ')
       

        #Find the separator
      
        Bar = NewStr.find("|")
       

        #Split into two strings
        PartA = NewStr[0:Bar]
       
        PartB=NewStr[Bar+1:len(NewStr)]
       

        #Strip out any leading or trailing spaces
        PartA=PartA.strip() 
        PartB=PartB.strip() 
       
        
        #create the lists
        ListA = list(PartA.split(" "))
       

        ListB = list(PartB.split(" "))
       

        CardList = []
        count=0
        Pts=0
        CardPts=0
        PrevPts = 0



        for element in ListA:
            if element in ListB:
                if element != '':
                    CardList.append(element)


        print(CardList)

        for i in CardList:  
            count = count+1
            
        if len(CardList) > 0:
            CardPts= 2**(count-1)
      
        print("Points",CardPts)
        TotalPts = TotalPts + CardPts
        print("TotalPts",TotalPts)

    
```
